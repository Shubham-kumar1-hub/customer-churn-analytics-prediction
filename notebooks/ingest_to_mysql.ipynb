```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("../data/Customer_Churn.csv")
```

```python
# Preview
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 3 rows:")
print(df.head(3))
print("\nData types:")
print(df.dtypes)
```

```python
# Fixing the TotalCharges column to numeric, coercing errors to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Nan Values that got created
print("Null values in TotalCharges:", df['TotalCharges'].isnull().sum())
print("Total nulls across all columns:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Dropping the 11 null rows -> they're new customers, not useful for churn modeling
df = df.dropna(subset=['TotalCharges'])

print("Shape after dropping nulls:", df.shape)

# Connecting to mySQL database
engine = create_engine("mysql+pymysql://root:12345@localhost:3306/churn_db")

df.to_sql(
    name='customers',
    con=engine,
    if_exists='replace',
    index=False
)

print("Data successfully loaded into MySQL!")
print("Total rows inserted:", len(df))
```
